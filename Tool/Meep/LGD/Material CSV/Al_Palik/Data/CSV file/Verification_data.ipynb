```python
import numpy as np
import matplotlib.pyplot as plt

# ------------------------------------------------------
# 1. CSV 파일에서 데이터 읽어오기
# ------------------------------------------------------
# 파일 포맷(3개 열): wavelength, real(n), imag(n)
filename = "Al_palik_data.csv"  # 예시 파일명
data = np.genfromtxt(filename, delimiter=",", skip_header=0)

# CSV 파일의 각 열을 변수에 할당
#   - 1열: wavelength (예: nm 또는 µm 등, 단위는 CSV에 맞춰 확인하세요.)
#   - 2열: real(n)
#   - 3열: imag(n)
wavelength = data[:, 0]
n_real = data[:, 1]
n_imag = data[:, 2]

# ------------------------------------------------------
# 2. 복소 굴절률, 복소 유전율 계산
# ------------------------------------------------------
n_complex = n_real + 1j * n_imag  # n + i*k
epsilon = n_complex**2           # (n + i*k)^2

# ------------------------------------------------------
# 3. 그래프 그리기 (실수부, 허수부)
# ------------------------------------------------------
fig, ax = plt.subplots(ncols=2, figsize=(10, 4))

# 실수부
ax[0].plot(wavelength, np.real(epsilon), 'b-o', label='Re($\\varepsilon$)')
ax[0].set_xlabel("Wavelength")
ax[0].set_ylabel("Re($\\varepsilon$)")
ax[0].legend()

# 허수부
ax[1].plot(wavelength, np.imag(epsilon), 'r-o', label='Im($\\varepsilon$)')
ax[1].set_xlabel("Wavelength")
ax[1].set_ylabel("Im($\\varepsilon$)")
ax[1].legend()

fig.suptitle("Dielectric Function vs. Wavelength", fontsize=14)

# ------------------------------------------------------
# 4. 그래프 저장
# ------------------------------------------------------
plt.tight_layout()
plt.savefig("epsilon_plot.png", dpi=150)  # 이미지 파일로 저장
plt.show()
```
